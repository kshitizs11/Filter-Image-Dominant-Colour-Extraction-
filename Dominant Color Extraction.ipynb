```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("./Kshitiz.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
x = img.reshape((-1,3))
```

```python
x.shape
```

```python
from sklearn.cluster import KMeans
```

```python
k=10
km = KMeans(n_clusters=k)
```

```python
km.fit(x)
```

```python
km.cluster_centers_
```

```python
centers = np.array(km.cluster_centers_, dtype="uint")
centers
```

Plot The Extracted Colors

```python
for c in centers:
    a = np.zeros((10,10,3),dtype = "uint")
    a[:,:,:] = c
    plt.imshow(a)
    plt.axis("off")
    plt.show()
```

```python
km.labels_
```

```python
new_img = np.zeros((x.shape[0],3), dtype = "uint")
```

```python
new_img.shape
```

```python
for i in range(x.shape[0]):
    new_img[i] = centers[km.labels_[i]]
```

```python
new_img.shape
```

```python
new_img = new_img.reshape((1344,1344,3))
```

```python
plt.imshow(new_img)
plt.axis("off")
plt.show()
```
